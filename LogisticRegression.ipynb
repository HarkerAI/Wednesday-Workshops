Some examples might be pass/fail, accepted/rejected, makes purchase/doesn't, and has disease/doesn't.

The basic idea is the we are trying to find the coefficient of x and a in the first equation, and then the last expression can be calculated for each input.

## Log Reg
We will use another sklearn package linear_model.LogisticRegression to perform the fit to get the intercept, a, and the coefficient, b.

```python
from sklearn import linear_model
X = data[['NumWhiskers']]
y = data[['Adopted']]
result = linear_model.LogisticRegression(C=100).fit(X, y.values.ravel()) #we have to use .values.ravel() 
                                                                         #to avoid dimension errors
print(result.coef_)
print(result.intercept_)
```

## Graphing
To visualize the regression, we plot the sigmoid function for 200 values of NumWhiskers between 0 and 30 along with the original data points.

```python
import numpy as np

a = result.intercept_
b = result.coef_

#Replot the original datapoints
data.plot(kind='scatter',x='NumWhiskers',y='Adopted',color='red')
plt.scatter(X, y, color='black', zorder=20)

# and plot the result
x = np.linspace(0, 30, 200) #creates range

def sigmoid(n):
    return 1 / (1 + np.exp(-n))   #sigmoid is another name for the log graph

pred = sigmoid(a + x * b).ravel()
plt.xlim(0, 30)
plt.plot(x, pred, color='red', linewidth=3)
```
